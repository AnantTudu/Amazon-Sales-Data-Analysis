## Data Analysis

```python
sales_data = df

# Display basic summary statistics
summary_statistics = sales_data[['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']].describe()
print("Summary Statistics:\n", summary_statistics)
```

```python
# Calculate additional statistics: median and mode
additional_stats = {}
for column in ['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']:
    median = sales_data[column].median()
    mode = sales_data[column].mode()[0]  # mode() returns a Series, take the first value
    additional_stats[column] = {'median': median, 'mode': mode}
```

```python
# Display additional statistics
additional_stats_df = pd.DataFrame(additional_stats).T
print("\nAdditional Statistics:\n", additional_stats_df)
```

```python
# Calculate standard deviation separately (already included in .describe())
standard_deviation = sales_data[['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']].std()
print("\nStandard Deviation:\n", standard_deviation)
```

```python
# Convert 'Order Date' to datetime, infer format
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'], infer_datetime_format=True, errors='coerce')

# Drop rows with NaT in 'Order Date' if any
sales_data = sales_data.dropna(subset=['Order Date'])

# Extract year and month
sales_data['Year'] = sales_data['Order Date'].dt.year
sales_data['Month'] = sales_data['Order Date'].dt.month

# Calculate total revenue by region
region_sales = sales_data.groupby('Region')['Total Revenue'].sum().sort_values(ascending=False)

# Print total revenue by region
print("Total Revenue by Region:\n", region_sales)

# Calculate total revenue by country
country_sales = sales_data.groupby('Country')['Total Revenue'].sum().sort_values(ascending=False)

# Print total revenue by country (Top 10 countries)
print("\nTop 10 Countries by Total Revenue:\n", country_sales.head(10))
```

```python
# Calculate total profit by item type without resetting the index
item_sales = sales_data.groupby('Item Type')['Total Profit'].sum().sort_values(ascending=False)
# Print the total profit by item type
print("Total Profit by Item Type:\n", item_sales)
```

```python
# Calculate total revenue and total profit by sales channel
channel_sales = sales_data.groupby('Sales Channel')[['Total Revenue', 'Total Profit']].sum().reset_index()
# If you want to show the data without visualization, you can print it
print("Channel Sales Data:\n", channel_sales)
```

```python
priority_sales = sales_data.groupby('Order Priority')[['Total Revenue', 'Total Profit']].sum().reset_index().sort_values(by='Total Revenue', ascending=False)

# Print the data 
print("Priority Sales Data:\n", priority_sales.to_string(index=False))
```

```python
# Calculate the correlation matrix
correlation_matrix = sales_data[['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']].corr()

# Print the correlation matrix
print("Correlation Matrix:\n", correlation_matrix)
```

```python
# Try inferring the date format
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'], errors='coerce')

# Check for any remaining NaT values if needed
na_dates = sales_data['Order Date'][sales_data['Order Date'].isna()]
if not na_dates.empty:
    print(f"Found {len(na_dates)} invalid date strings: {na_dates}")

# Extract year and month from 'Order Date'
sales_data['Year'] = sales_data['Order Date'].dt.year
sales_data['Month'] = sales_data['Order Date'].dt.month

# Calculate monthly sales
monthly_sales = sales_data.groupby(['Year', 'Month'])['Total Revenue'].sum().reset_index()
print("Monthly Sales:\n", monthly_sales)

# Calculate yearly sales
yearly_sales = sales_data.groupby('Year')['Total Revenue'].sum().reset_index()
print("\nYearly Sales:\n", yearly_sales)
```

## Data Visualisation

```python
# Assuming sales_data is already loaded as a DataFrame

# Calculate total profit by item type without resetting the index
item_sales = sales_data.groupby('Item Type')['Total Profit'].sum().sort_values(ascending=False)


# Calculate additional statistics: median and mode
additional_stats = {}
for column in ['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']:
    median = sales_data[column].median()
    mode = sales_data[column].mode()[0]  # mode() returns a Series, take the first value
    additional_stats[column] = {'median': median, 'mode': mode}

# Display additional statistics
additional_stats_df = pd.DataFrame(additional_stats).T

# Plot median and mode for additional statistics
additional_stats_df.plot(kind='bar', figsize=(14, 8))
plt.title('Median and Mode of Various Metrics')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.show()

# Calculate standard deviation separately (already included in .describe())
standard_deviation = sales_data[['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']].std()

# Plot standard deviation for various metrics
plt.figure(figsize=(12, 6))
sns.barplot(x=standard_deviation.values, y=standard_deviation.index, palette='viridis')
plt.title('Standard Deviation of Various Metrics')
plt.xlabel('Standard Deviation')
plt.ylabel('Metrics')
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(data=monthly_sales, x='Month', y='Total Revenue', hue='Year', marker='o')
plt.title('Monthly Sales Trend')
plt.show()

# Yearly sales trend
yearly_sales = sales_data.groupby('Year')['Total Revenue'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=yearly_sales, x='Year', y='Total Revenue', palette='viridis')
plt.title('Yearly Sales Trend')
plt.show()
```

```python


numerical_columns = ['Units Sold', 'Unit Price', 'Unit Cost', 'Total Revenue', 'Total Cost', 'Total Profit']

for col in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(sales_data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.boxplot(x=sales_data[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python
region_sales = sales_data.groupby('Region')['Total Revenue'].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=region_sales, x='Region', y='Total Revenue', palette='viridis')
plt.title('Total Revenue by Region')
plt.xticks(rotation=90) 
plt.show()

country_sales = sales_data.groupby('Country')['Total Revenue'].sum().reset_index().sort_values(by='Total Revenue', ascending=False)

plt.figure(figsize=(14, 8))
sns.barplot(data=country_sales.head(10), x='Country', y='Total Revenue', palette='viridis')
plt.title('Top 10 Countries by Total Revenue')
plt.show()
```

```python
# Calculate total profit by item type without resetting the index
item_sales = sales_data.groupby('Item Type')['Total Profit'].sum().sort_values(ascending=False)

# If you still want to visualize the data using a bar plot
plt.figure(figsize=(14, 8))
sns.barplot(x=item_sales.values, y=item_sales.index, palette='viridis')
plt.title('Total Profit by Item Type')
plt.xlabel('Total Profit')
plt.ylabel('Item Type')
plt.show()
```

```python
# Calculate total revenue and total profit by sales channel
channel_sales = sales_data.groupby('Sales Channel')[['Total Revenue', 'Total Profit']].sum().reset_index()

# Plot total revenue by sales channel
plt.figure(figsize=(12, 6))
sns.barplot(data=channel_sales, x='Sales Channel', y='Total Revenue', palette='viridis')
plt.title('Total Revenue by Sales Channel')
plt.show()

# Plot total profit by sales channel
plt.figure(figsize=(12, 6))
sns.barplot(data=channel_sales, x='Sales Channel', y='Total Profit', palette='viridis')
plt.title('Total Profit by Sales Channel')
plt.show()

```

```python
priority_sales = sales_data.groupby('Order Priority')[['Total Revenue', 'Total Profit']].sum().reset_index().sort_values(by='Total Revenue', ascending=False)

# Plot total revenue by order priority
plt.figure(figsize=(12, 6))
sns.barplot(data=priority_sales, x='Order Priority', y='Total Revenue', palette='viridis')
plt.title('Total Revenue by Order Priority')
plt.show()

# Plot total profit by order priority
plt.figure(figsize=(12, 6))
sns.barplot(data=priority_sales, x='Order Priority', y='Total Profit', palette='viridis')
plt.title('Total Profit by Order Priority')
plt.show()

```

```python
# Plot the correlation matrix as a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True, linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
sales_data['Ship Date'] = pd.to_datetime(sales_data['Ship Date'])
sales_data['Shipping Time'] = (sales_data['Ship Date'] - sales_data['Order Date']).dt.days

plt.figure(figsize=(12, 6))
sns.histplot(sales_data['Shipping Time'], kde=True)
plt.title('Distribution of Shipping Time')
plt.show()

plt.figure(figsize=(12, 6))
sns.boxplot(x=sales_data['Shipping Time'])
plt.title('Boxplot of Shipping Time')
plt.show()

```
